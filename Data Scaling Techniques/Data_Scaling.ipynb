```python
#Importing pandas for data handling
import pandas as pd

#Importing numpy for numerical computations
import numpy as np
```

**1. Min_Max Scaler**

**Import min_max scaler (normalization_scaler)**

```python
#Importing MinMaxScaler for normalizing features within a fixed range
from sklearn.preprocessing import MinMaxScaler
```

**Define** **Data**

```python
#Creating a DataFrame with sample values for columns X and Y
data = pd.DataFrame(data=[[5,5], [10,2], [2,0], [0,10]], columns=['X', 'Y'])
data  #Displaying the dataset
```

**Plot Scatter Plot**

```python
import matplotlib.pyplot as plt
plt.scatter(data.X, data.Y)      #Plotting a scatter plot for columns X and Y
plt.axvline(x=0, c='red')        #Adding a vertical line at x = 0
plt.axhline(y=0, c='yellow')     #Adding a horizontal line at y = 0
plt.show()       #Displaying the plot
```

**Define, fit and transform min-max scaler**

```python
#Initializing the MinMaxScaler
scaler = MinMaxScaler()       #y=(x-min)/(max-min)
min_max_scaled = scaler.fit_transform(data)  #Fitting the scaler on data and transforming it to scaled values
```

**Print Scaled Data**

```python
#Converting the scaled NumPy array back to a DataFrame with column names
min_max_scaled = pd.DataFrame(min_max_scaled, columns=['X', 'Y'])
min_max_scaled   #Displaying the scaled dataset
```

**Plot both min_max Scaler**

```python
import matplotlib.pyplot as plt
plt.scatter(min_max_scaled.X, min_max_scaled.Y) #Plotting the scatter plot for scaled X and Y values
plt.axvline(x=0, c='red')     #Adding a vertical line at x = 0
plt.axhline(y=0, c='yellow')  #Adding a horizontal line at y = 0
plt.show()   #Displaying the plot
```

**Subplots**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))   # Creating a figure with 2 subplots (side by side)

#Left plot: Original data
plt.subplot(1,2,1)
plt.scatter(data.X, data.Y)
plt.axvline(x=0, c='red')
plt.axhline(y=0, c='yellow')

#Right plot: Scaled data
plt.subplot(1,2,2)
plt.scatter(min_max_scaled.X, min_max_scaled.Y)
plt.show()    #Displaying the plots
```

**2. Standar Scaler**

**Import Standard Scaler(standardization_scaler)**

```python
#Importing StandardScaler for standardizing features
from sklearn.preprocessing import StandardScaler
```

```python
#Creating the same sample dataset with columns X and Y
data = pd.DataFrame(data=[[5,5], [10,2], [2,0], [0,10]], columns=['X', 'Y'])
data  #Displaying the dataset
```

```python
import matplotlib.pyplot as plt
plt.scatter(data.X, data.Y)      #Scatter plot of original X and Y values
#Adding reference lines at x = 0 and y = 0
plt.axvline(x=0, c='red')
plt.axhline(y=0, c='yellow')
#Displaying the plot
plt.show()
```

```python
scaler1=StandardScaler()   #Initializing the StandardScaler
std_scaled=scaler1.fit_transform(data) #Fitting the scaler on the dataset and transforming it
std_scaled  #Displaying the standardized values (NumPy array)
```

```python
#Converting the standardized NumPy array back into a DataFrame
std_scaled = pd.DataFrame(std_scaled, columns=['X', 'Y'])
std_scaled  #Displaying the standardized dataset
```

```python
std_scaled.describe() #Displaying summary statistics of the standardized dataset
#Mean should be ~0 and standard deviation ~1 for each column
```

**Plot Standard Scaler**

```python
import matplotlib.pyplot as plt
#Scatter plot of standardized X and Y values
plt.figure(figsize=(5,5))
plt.scatter(std_scaled.X, std_scaled.Y)

#Adding reference lines at x = 0 and y = 0
plt.axvline(x=0, c='red')
plt.axhline(y=0, c='yellow')

#Displaying the plot
plt.show()
```

**Subplots**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5)) #Creating a figure with 2 subplots (side by side) for comparison

#Left subplot: Original data
plt.subplot(1,2,1)
plt.scatter(data.X, data.Y)
plt.axvline(x=0, c='red')
plt.axhline(y=0, c='yellow')

#Right subplot: Standard scaled data
plt.subplot(1,2,2)
plt.scatter(std_scaled.X, std_scaled.Y)
plt.axvline(x=0, c='red')
plt.axhline(y=0, c='yellow')

#Displaying both plots
plt.show()
```

**Plot both min_max and Standard Scaler**

```python
import matplotlib.pyplot as plt
#Scatter plot overlaying Min-Max scaled and Standard scaled data
plt.scatter(min_max_scaled.X, min_max_scaled.Y, marker ='o')  #Min-Max scaled points: circle marker 'o'
plt.scatter(std_scaled.X, std_scaled.Y, marker ='^') #Standard scaled points: triangle marker '^'
# Adding reference lines at x = 0 and y = 0
plt.axvline(x=0, c='red')
plt.axhline(y=0, c='yellow')
plt.show() # Displaying the overlay plot
```
