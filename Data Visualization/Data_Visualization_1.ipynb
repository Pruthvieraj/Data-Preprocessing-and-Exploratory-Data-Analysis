```python
#ax.bar(x, height, width, bottom, align)
#Import the required library
import matplotlib.pyplot as plt

#Create a figure object
fig = plt.figure()

#Add axes to the figure: [left, bottom, width, height]
ax = fig.add_axes([0, 0, 1, 1])

#Define categories (programming languages) and their corresponding values
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 17, 35, 29, 12]

#Create a bar plot
ax.bar(langs, students)

#Display the plot
plt.show()
```

```python
#Import required libraries
import numpy as np
import matplotlib.pyplot as plt

#Each list represents values for different categories (3 datasets, each with 4 values)
data = ([30, 25, 50, 20],
        [40, 23, 51, 17],
        [35, 22, 45, 19])

#X locations for the groups (4 categories)
X = np.arange(4)

#Create a figure object
fig = plt.figure()

#Add axes to the figure: [left, bottom, width, height]
ax = fig.add_axes([0, 0, 1, 1])

#Plot bars for each dataset, shifting each by 0.25
ax.bar(X + 0.00, data[0], color='b', width=0.25, label='Dataset 1')
ax.bar(X + 0.25, data[1], color='g', width=0.25, label='Dataset 2')
ax.bar(X + 0.50, data[2], color='r', width=0.25, label='Dataset 3')

#Add legend for clarity
ax.legend()

#Display the plot
plt.show()
```

```python
#Import required libraries
import numpy as np
import matplotlib.pyplot as plt

#Number of groups
N = 5

#Scores for men and women
menMeanscore = (20, 35, 30, 35, 27)
womenMeanscore = (25, 32, 34, 20, 25)

#X locations for the groups
ind = np.arange(N)

#Width of each bar
width = 0.35

#Create a figure object
fig = plt.figure()

#Add axes to the figure: [left, bottom, width, height]
ax = fig.add_axes([0, 0, 1, 1])

#Plot stacked bars
ax.bar(ind, menMeanscore, width, color='r')
ax.bar(ind, womenMeanscore, width, bottom=menMeanscore, color='g')

#Set y-axis label
ax.set_ylabel('Scores')

ax.set_title('Scores by group and gender') #Set plot title

#Set x-axis ticks and labels
ax.set_xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
ax.set_yticks(np.arange(0, 81, 10))  #Set y-axis ticks

ax.legend(('Men', 'Women'))  #Add legend

plt.show()  #Display the plot
```

```python
#Import required libraries
from matplotlib import pyplot as plt
import numpy as np
fig, ax = plt.subplots(1, 1) #Create a figure and axes
#Data to plot
a = np.array([22, 87, 5, 43, 9, 73, 55, 54, 11, 20, 51, 5, 79, 31, 27])
#Plot histogram with specified bins
ax.hist(a, bins=[0, 25, 50, 75, 100])
ax.set_title("Histogram of Result")  #Set plot title
ax.set_xticks([0, 25, 50, 75, 100])   #Set x-axis ticks
#Set axis labels
ax.set_xlabel('Marks')
ax.set_ylabel('Number of Students')
plt.show()  #Display the plot
```
