**KNN FOR QUANTITATIVE OR NUMERICAL DATA**

**Importing Essential Libraries**

```python
import numpy as np    #Import NumPy for numerical computations
import pandas as pd     #Import pandas for data handling and analysis
```

**Importing the KNNimputer**

```python
from sklearn.impute import KNNImputer    #Import KNNImputer for filling missing values using nearest neighbors
```

**Create dataset for marks of students**

```python
#Create a dictionary of student scores with some missing values
dict = {'Maths' : [85, 91, np.nan, 91],
        'Chemistry' : [72, 75, 74, np.nan],
        'Physics' : [np.nan, 68, 82, 81],
        'Biology' : [81, 81, 72, np.nan]
        }
```

**Creating a dataframe from the dictionary**

```python
#Convert dictionary to DataFrame to visualize scores with missing data
before_imputation = pd.DataFrame(dict)

#Display the DataFrame
before_imputation
```

**Replacing the missing values with KNNimputer**

```python
imputer= KNNImputer(n_neighbors=10)   #Initialize KNN imputer with 10 neighbors
after_imputation = imputer.fit_transform(before_imputation)   #Fill missing values in the DataFrame using KNN imputation
after_imputation    #Show the imputed data as a NumPy array
```

**Convert the array into dataframe**

```python
before_imputation.columns  #Get the list of column names in the DataFrame
```

```python
after_imputation = pd.DataFrame(after_imputation, columns=before_imputation.columns)   # Convert imputed NumPy array back to DataFrame with original column names
after_imputation     #Display the DataFrame after imputation
```
