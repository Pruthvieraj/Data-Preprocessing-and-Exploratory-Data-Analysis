**Filling Missing Values with a Specific Constant**

```python
df.fillna(10)  #Replace all missing values in the DataFrame with 10
```

```python
df['age']   #Actual Column
```

**Filling Missing Values in a Specific Column**

```python
# Replace missing values in the 'age' column with 50
df['age'] = df['age'].fillna(50)
df['age']
```

```python
#Replace missing values in 'first_name' column with 'Kashish'
df['first_name'] = df['first_name'].fillna('Kashish')

df['first_name']      #Display the updated 'first_name' column
```

```python
df['last_name']=df['last_name'].fillna('Chelwani')  #Replace missing values in 'last_name' column with 'Chelwani'
df['last_name']       #Show the updated 'last_name' column
```

**Data Imputation**

```python
df['pre_movie_score']    #Access the 'pre_movie_score' column from the DataFrame
```

```python
# Calculate the average value of the 'pre_movie_score' column,ignoring missing values
df['pre_movie_score'].mean()
```

```python
#Fill missing 'pre_movie_score' values with the column's average score
df['pre_movie_score'] = df['pre_movie_score'].fillna(df['pre_movie_score'].mean())

df['pre_movie_score']       #Display the updated 'pre_movie_score' column
```

```python
#Fill missing 'post_movie_score' values with the column's median score
df['post_movie_score'] = df['post_movie_score'].fillna(df['post_movie_score'].median())

#Display the DataFrame after filling missing values
df
```
