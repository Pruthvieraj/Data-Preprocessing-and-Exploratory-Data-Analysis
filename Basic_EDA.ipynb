**Basic operations with a dataframe**

**Mathematical Operations**

```python
#Mathematical Operations
result = stats.BirthRate * stats.InternetUsers  #Multiplying 'BirthRate' and 'InternetUsers' columns element-wise
print(result.head())  #Displaying the first 5 results
```

```python
#Add column
stats['MyCalc'] = stats.BirthRate * stats.InternetUsers #Creating a new column 'MyCalc' as the product of BirthRate and InternetUsers
stats.head()  #Displaying the first 5 rows to verify the new column
```

```python
#removing a column
#axis=1 indicates column, and assignment ensures the change is saved
stats = stats.drop('MyCalc', axis=1)         #create a new object by default unless you assign it
stats.head()   #Displaying the first 5 rows to confirm the column has been removed
```

**Filtering Dataframes**

```python
filter = stats['InternetUsers'] < 2  #Creating a filter for countries where InternetUsers < 2%
filter  # Displaying the Boolean filter
#stats[filter]- to see only the rows that meet the condition
```

```python
filter = stats['InternetUsers'] < 2 #Creating a filter for countries where InternetUsers < 2%
filter
stats[filter] #Using the filter to display only rows that meet the condition
```

```python
#filter by BirthRate
Filter2 = stats.BirthRate > 40
stats[Filter2]  #display only rows that meet the condition
```

```python
#could also combine into one step
stats[stats.BirthRate > 40]   #create filter and pass it to stats brackets
```

```python
#applying two filters
stats[filter & Filter2]
```

```python
#applying the filters using squre brackets approach
stats[(stats.BirthRate>40)&(stats.InternetUsers<2)]
```
