```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({
    'team': ['A', 'A', 'B', 'B', 'B', 'B', 'C', 'C'],
    'points': [25, 12, 15, 14, 19, 23, 25, 29]})

print(df)
```

**Perform one-hot encoding**

```python
from sklearn.preprocessing import OneHotEncoder

#creating instance of one-hot-encoder
encoder=OneHotEncoder(handle_unknown='ignore')

#perform one-hot encoding on 'team' column
encoder_df = pd.DataFrame(encoder.fit_transform(df[['team']]).toarray())

#merge one-hot encoded columns back with original DataFrame
final_df = df.join(encoder_df)

#view final df
print(final_df)
```

```python
#drop 'team' variable
final_df.drop('team', axis=1, inplace=True)

#view final df
print(final_df)
```


**Practice Example**

Create a dataframe with column as 'Gender, 'Age', 'Degree'.
Create 10 data values in this data set
Apply one-hot encoding on the columns
1. Gender
2. Degree

```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Female', 'Male'],
    'Age': [1, 21, 42, 31, 19, 17, 41, 3, 3, 21],
    'Degree': [1, 2, 1, 3, 2, 1, 2, 1, 1, 2]
})

print(df)
```

```python
from sklearn.preprocessing import OneHotEncoder

#creating instance of one-hot-encoder
encoder1=OneHotEncoder(handle_unknown='ignore')

#perform one-hot encoding on 'gender' column
encoder_df = pd.DataFrame(encoder1.fit_transform(df[['Gender']]).toarray())

#merge one-hot encoded columns back with original DataFrame
final_df1 = df.join(encoder_df)

#view final df
print(final_df1)
```

```python
from sklearn.preprocessing import OneHotEncoder

#creating instance of one-hot-encoder
encoder1=OneHotEncoder(handle_unknown='ignore')

#perform one-hot encoding on 'degree' column
encoder_df = pd.DataFrame(encoder1.fit_transform(df[['Degree']]).toarray())

#merge one-hot encoded columns back with original DataFrame
final_df1 = df.join(encoder_df)

#view final df
print(final_df1)
```

```python
final_df1.drop(['Gender','Degree'],axis=1, inplace=True) ## Drop Original Gender and Degree Columns After Encoding

print(final_df1)
```
