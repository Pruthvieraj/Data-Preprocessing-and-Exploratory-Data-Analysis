**Simple Method of Imputing Age Column**

```python
#fill in missing values in age column
age_mean_values=df['Age'].mean()        #Calculate the mean age
df['Age']=df['Age'].fillna(age_mean_values)          #Fill missing values in the 'Age' column with the mean age
df
```

```python
df.isna().sum()        #Checking for any missing values
```

```python
#Fill missing values in 'Embarked' using forward fill
df['Embarked'].fillna(method='ffill', inplace=True)   #Copies last valid value forward
df.isna().sum()       #Check again for missing values after filling
```

**Advanced method of Imputing Age column**

```python
#create new title column
df['Title']=df['Name'].str.extract('([A-Za-z]+)\.', expand=True)     #Extracting titles from 'Name'
df['Title'].value_counts() #Count the frequency of each title
```

```python
df.head()      #Check recent updates
```

```python
#replace rare titles with more common ones
mapping = {'Mlle':'Miss', 'Major' : 'Mr', 'Col':'Mr', 'Sir':'Mr', 'Don':'Mr', 'Mme':'Mrs', 'Jonkheer':'Mr', 'Lady':'Mrs', 'Capt':'Mr', 'Countess':'Mrs', 'Ms':'Miss', 'Dona':'Mrs'}
df.replace({'Title':mapping}, inplace=True)  #With inplace=True changes are applied directly to the original DataFrame
```

```python
df['Title'].value_counts()    #Count the frequency of each title after replacing rare ones
```

```python
df.groupby('Title')['Age'].median()    #Calculate the median age for each group
```

```python
#impute missing Age values using median of Title Groups
title_ages=dict(df.groupby('Title')['Age'].median())   #Create a dictionary with median ages for each title
title_ages
```

```python
#Before Imputation
df.head()
```

```python
#create a column of average ages by mapping each Title to its median age
df['Age_med'] = df['Title'].apply(lambda x: title_ages[x])
df #After Imputation
```
