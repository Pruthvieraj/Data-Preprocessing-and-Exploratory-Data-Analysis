```python
#importing the modules
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#generating 2D 10X10 matrix of random numbersmfrom 1 to 100
data = np.random.randint(low=1, high=100, size=(10,10))
print("The data to be plotted:\n")
print(data)
#plotting the heatmap
hm=sns.heatmap(data=data)
#displaying the potted heatmap
plt.show()
```

**Displaying the cell values**

```python
#generating 2D 10X10 matrix of random numbers from 1 to 100
data = np.random.randint(low=1, high=100, size=(10,10))
#setting the parameter values
annot=True
#plotting the heatmap
hm=sns.heatmap(data=data, annot=annot)
#displaying the potted heatmap
plt.show()
```

```python
#generating 2D 10X10 matrix of random numbers from 1 to 100
data = np.random.randint(low=1, high=100, size=(10,10))
#setting the parameter value
cmap='tab20'
#plotting the heatmap
hm=sns.heatmap(data=data, cmap=cmap)
#displaying the potted heatmap
plt.show()
```

**customizing the separting lines**

```python
#generating 2D 10X10 matrix of random numbers from 1 to 100
data = np.random.randint(low=1, high=100, size=(10,10))
#setting the parameter value
linewidths=2
linecolor='yellow'
#plotting the heatmap
hm=sns.heatmap(data=data, linewidths=linewidths,linecolor=linecolor)
#displaying the potted heatmap
plt.show()
```

**Hiding the colorbar**

```python
#generating 2D 10X10 matrix of random numbers from 1 to 100
data = np.random.randint(low=1, high=100, size=(10,10))
#setting the paramter values
cbar=False
#plotting the heatmap
hm=sns.heatmap(data=data, cbar=cbar)
#displaying the potted heatmap
plt.show()
```

```python
#removing the labels
#importing the modules
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#generating 2D 10X10 matrix of random numbers from 1 to 100
data = np.random.randint(low=1, high=100, size=(10,10))
#setting parameter values
xticklabels=False
yticklabels=False
#plotting the heatmap
hm=sns.heatmap(data=data, xticklabels=xticklabels, yticklabels=yticklabels)
#displaying the potted heatmap
plt.show()
```

```python
tips = sns.load_dataset("tips")  #Loading tips dataset
tips.head()   #Display the 1st 5 rows
```

```python
#matrix from for correlation data
tips.corr(numeric_only=True)
```

```python
sns.heatmap(tips.corr(numeric_only=True)) #Generating a heatmap
```

```python
#Create a heatmap of correlation matrix for numeric columns
#annot=True displays the numerical values on the heatmap squares
sns.heatmap(tips.corr(numeric_only=True), cmap='coolwarm', annot=True)
```

```python
flights=sns.load_dataset('flights')  #Loading the flights Dataset
flights.head()
```

```python
#Create a pivot table with months as rows and years as columns
#Values are the number of passengers
flights.pivot_table(values='passengers', index='month', columns='year')
```

```python
#Create pivot table with months as rows and years as columns
pvflights = flights.pivot_table(values='passengers', index='month', columns='year')

#Create a heatmap of the pivot table
sns.heatmap(pvflights)
```

```python
#Create a heatmap with custom color map and grid lines
#cmap='magma' sets the color scheme
#linecolor='white' sets color of lines between cells
#linewidths=1 sets the thickness of the lines
sns.heatmap(pvflights, cmap='magma', linecolor='white', linewidths=1)
```
