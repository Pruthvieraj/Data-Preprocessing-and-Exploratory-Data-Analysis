```python

import matplotlib.pyplot as plt
import numpy as np
fig, axes = plt.subplots(1, 3, figsize=(12, 4))  #3 subplots arranged in 1 row
x = np.arange(1, 11)
#Plot cubic curve with default grid
axes[0].plot(x, x**3, 'g', lw=2)
axes[0].grid(True)
axes[0].set_title('default grid')

#exponential curve with customized grid
axes[1].plot(x, np.exp(x), 'r')
axes[1].grid(color='b', ls='--', lw=0.25)
axes[1].set_title('custom grid')

axes[2].plot(x, x)
axes[2].set_title('no grid')
fig.tight_layout()
plt.show()
```

```python
#Import library
import matplotlib.pyplot as plt
#Create figure and axes
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
#Enable grid
ax.grid(True)
#Customize spines (borders) of the plot
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_color('red')
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_color('black')
ax.spines['top'].set_color('green')
#Plot sample data
ax.plot([5, 6, 7, 8, 1])
plt.show()
```

```python
#Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#Load the Iris dataset
iris = sns.load_dataset('iris')

#Display first 5 rows
iris.head()
```

```python
#just the grid
sns.pairplot(data = iris)
```

```python
#Create a pair plot and color points based on species
#hue='species' differentiates species with colors
#palette='rainbow' sets the color scheme
sns.pairplot(iris, hue='species', palette='rainbow')
```

**pairgrid mechanism**

```python
#Initialize a PairGrid object for the iris dataset
sns.PairGrid(data=iris)
```

```python
#then you map to the grid, pairgrid
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)  #Initialize PairGrid for iris dataset
g.map_diag(plt.hist)  #Map histogram to the diagonal
g.map_upper(plt.scatter)  #Map scatter plots to the upper triangle
g.map_lower(sns.kdeplot)  #Map KDE plots to the lower triangle
```

```python
#using tips dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
#just the grid
g = sns.FacetGrid(tips, col = 'time', row = 'smoker')
```

```python
#the total bills are plotted as histogram across the grids we created using 'map'
#Visualize the distribution of total_bill for different times of day and smoking status in the tips dataset using a grid layout.
g = sns.FacetGrid(tips, col='time', row='smoker')  #Create FacetGrid with columns=time and rows=smoker
g.map(plt.hist, 'total_bill')  #Map histogram of total_bill onto the grid
```

```python
#over here we have defined hue as sex
#and also plotted a scater plot where x axis is total bill and y axis is tip
g = sns.FacetGrid(tips, col = 'time', row = 'smoker', hue = 'sex', palette='coolwarm')
g = g.map(plt.scatter, 'total_bill', 'tip').add_legend()
```

**Joint** **grid**

```python
g = sns.JointGrid(x='total_bill', y='tip', data=tips)  #Create a JointGrid for total_bill vs tip\
```

```python
g = sns.JointGrid(x='total_bill', y='tip', data=tips)  #Create a JointGrid for total_bill vs tip
g = g.plot(sns.regplot, sns.distplot)  #Add regression plot on joint and distribution plots on margins
```

```python
#implementation of matplotlib function
import matplotlib.pyplot as plt
import numpy as np
#dummy data
x1 = np.linspace(0.0,5.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
#creates two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))
#plot without grid
ax1.plot(x1, y1)
ax1.set_title("Plot without Grid")
#plot with grid
ax2.plot(x1, y1)
ax2.set_title("Plot with Grid")
#draw gridlines
ax2.grid(True)
plt.show()
```

```python
#set line properties for the grid
#now lets draw gridlines using extra line properties such as color, linestyle and linewidth
#implementation of matplotlib function
import matplotlib.pyplot as plt
import numpy as np
#dummy data
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
#set graph color
plt.plot(x, y, 'green')
#to set title
plt.title("Plot with linewidth and linestyle")
#draw gridlines of grey color using given linewidth and linestyle
plt.grid(color='grey', linewidth = '0.4', linestyle=':')
plt.show()
```
