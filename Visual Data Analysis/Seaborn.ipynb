```python
#import seaborn
import seaborn as sns

#apply the default theme
sns.set_theme()

#load the iris dataset
tips = sns.load_dataset("tips")
```

```python
#Load the tips dataset again (already loaded above)
tips = sns.load_dataset("tips")
#Display the first few rows of the dataframe
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)     # Create a bar plot of tip amount by sex
```

```python
import numpy as np #import required libraries
```

```python
sns.countplot(x='day', data=tips) #Create a count plot of the number of tips by day
```

```python
# Create a box plot of total bill by day
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
#we can do entire dataframe with orient='h'
sns.boxplot(data=tips, palette='rainbow', orient='h')
#if orient not mentioned, always vertically plotted
```

```python
# Create a box plot of total bill by day, separated by sex using hue
#hue gives different colors with unique values of that category  mentioned
#how much max money is paid and by male or female?
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips, palette='coolwarm')
```

```python
#Compare the distribution of total_bill amounts for each day of the week, showing separate distributions for smokers and non-smokers using different colors
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='rainbow')
```

```python
# Create a violin plot of total bill by day
sns.violinplot(x='day', y='total_bill', data=tips, palette='rainbow') #the white line in plot shows the mean value.
```

```python
# Create a violin plot of total bill by day, separated by sex using hue
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, palette='Set1')
```

```python
# Create a distribution plot of total bill, separated by time (Lunch/Dinner)
sns.displot(data=tips, x='total_bill', col='time', kde=True)   #kernel density estimator
```

```python
# Create a distribution plot of total bill (deprecated function)
sns.distplot(tips['total_bill']) #for numerical components only
```

```python
sns.distplot(tips['total_bill'], kde=True, bins=5) # Create a distribution plot of total bill with 5 bins and KDE
```

```python
#Use kde=False to remove the KDE layer and show only histogram
sns.distplot(tips['total_bill'], kde=False, bins=5)
```

```python
#Create a joint plot showing relationship between total_bill and tip
#kind='scatter' displays a scatter plot
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
```

```python
#Create a joint plot with hexagonal bins
#kind='hex' shows density using hexagons
#color='red' sets the color of the hexagons
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex', color='red')
```

```python
#Create a joint plot with regression line
#kind='reg' fits a linear regression line to the data
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
#Create a pair plot for all numeric columns in the dataset
#Useful to visualize relationships and check colinearity
sns.pairplot(tips)
```

```python
#Create a pair plot and color points based on 'sex'
#hue='sex' differentiates male and female using colors
#palette='coolwarm' sets the color scheme
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
#Create a rug plot for the 'day' column
#Each tick represents an individual data point
sns.rugplot(tips['day'])
```

```python
#Create a Kernel Density Estimate plot for 'tip'
sns.kdeplot(tips['tip'])

#Add a rug plot on the same axis to show individual data points
sns.rugplot(tips['tip'])
```

```python
#Create a Kernel Density Estimate plot for 'total_bill'
sns.kdeplot(tips['total_bill'])

#Add a rug plot to show individual data points
sns.rugplot(tips['total_bill'])
```
